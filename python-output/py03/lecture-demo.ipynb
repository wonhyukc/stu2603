### 문자열 & 리스트 심화 메서드

```python
words = ['Python', 'is', 'Fun']
joined_text = " ".join(words)
print(f"Joined: {joined_text}")

print(f"'h' count: {'hello python'.count('h')}")

print(f"'Python' index: {words.index('Python')}")
```

### 리스트 기초 실습 (추가, 삭제, 복제)
변수 상자 안의 내용물을 넣고 빼는 기본 돔작입니다.

```python
stu01 = ["Geoffery Hinton", 95, 65, 90]
stu01.append(100) # 맨 끝에 추가
print(f"After append: {stu01}")

del stu01[0] # 이름 삭제
print(f"After del: {stu01}")

stu_new = stu01.copy()
print(f"Copied list: {stu_new}")
```

### f-string 소숫점 및 정렬
f-string을 통해 출력 혐태를 정교하게 조절합니다.

```python
val = 3.512345
print(f"[소숫점 두자리] {val:.2f}")

s = "Hello"
print(f"[오른쪽 정렬] {s:>10s}")
print(f"[가운데 정렬] {s:^10s}")
```

### 타입 변환 (Type Casting)

```python
num_str = "100"
num_int = int(num_str)
print(num_int + 50)  # 150 출력

num_float = float("3.14")
print(num_float * 2)

score = 95
print("내 점수는 " + str(score) + "점!")  # 문자열 합치기 시 str() 필요
```

# [S5] 문자열 & 자료구조 확장 (Advanced)

### 리스트 심화 기능
항목들을 덧붙이거나 끼워넣고, 정렬하는 방법입니다.

```python
colors = ["red", "blue", "green"]
colors.insert(1, "yellow")  # 1번 인덱스에 'yellow' 삽입
print(colors)

colors.extend(["black", "white"])  # 리스트 통째로 이어붙이기
print(colors)

colors.sort()  # 알파벳 순 정렬
print("Sorted:", colors)

popped_item = colors.pop()  # 맨 뒤 요소 뽑아내기
print("Popped item:", popped_item)
print("Remaining info:", colors)
```

### (참고) 딕셔너리(Dictionary) 맛보기
리스트가 순서대로 데이터를 보관한다면, 딕셔너리는 '키(Key)'와 '값(Value)'의 쌍으로 보관합니다.

```python
student_info = {
    "name": "Jane Doe",
    "age": 22,
    "major": "Computer Science"
}
print(student_info["name"])
student_info["grade"] = "A+"
print(student_info)
```
