```
# 리스트 생성
x = [3, 1, 2]

print(x)
print(x[1], x[2], x[0])
print(x[-1])
print(x[-2])
```

```
# 자르기(Slicing)

# range - 정수 목록을 만드는 함수
nums = list(range(10))
print(nums)
nums1 = list(range(2, 10))
print(nums1)
nums2 = list(range(2, 10, 2))
print(nums2)

# 자르기
print(nums[2:4])
print(nums[7:])
print(nums[:3])
print(nums[:-2])

# 잘라낸 부분을 새로운 리스트로 할당(assign)
nums[2:4] = [8, 9, 10]
print(nums)
```

```
# list 복사. 두 방식은 차이가 있다!
nums_copy_1 = nums
nums_copy_2 = nums[:]
nums_copy_3 = nums.copy()
print(nums_copy_1)
print(nums_copy_2)
print(nums_copy_3)
```

```
nums[0] = 'b'
print(nums)
print(nums_copy_1)
print(nums_copy_2)
print(nums_copy_3)
```

```
nums_copy_1[0] = 'a'
print(nums_copy_1)
print(nums_copy_2)
print(nums)
```

#### 반복문 with 리스트

```
# 반복문(Loops)
animals = ['cat', 'dog', 'monkey']
for animal in animals:
    print(animal)

animals = ['cat', 'dog', 'monkey']
for idx, animal in enumerate(animals):
    print('#%d: %s' % (idx + 1, animal))
```

#### list comprehension

```
# 반복문
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)
print(squares)
```

```
# 확장 문법 - 반복문
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]
print(squares)
```

```
# 확장 문법 - 반복문 + 조건문
nums = [0, 1, 2, 3, 4]
even_squares = [x ** 2 for x in nums if x % 2 == 0]
print(even_squares)  # Prints "[0, 4, 16]"
```

# 3.자료형(Container) - Sets

* 집합
* 중복된 값을 갖지 않는 리스트

```
animals = {'cat', 'dog'}
print(animals)

# 요소가 있는지 체크
print('cat' in animals)   # "True"
print('fish' in animals)  # "False"

# 요소 추가
animals.add('fish')
print('fish' in animals)  # "True"
print(animals)

# set 요소 갯수
print(len(animals))       # "3"
animals.add('cat')        # 중복된 값을 추가하면 무시. 리스트와 비교.
print(len(animals))       # "3"
print(animals)

# 요소 삭제
animals.remove('cat')
print(len(animals))       # "2"

######################
# List 비교
list_animals = ['cat', 'dog', 'fish']
print(list_animals)
print(len(list_animals))       # "3"

# list 요소 갯수
list_animals.append('cat')     # 중복해서 넣어도 추가된다.
print(len(list_animals))       # "4"
print(list_animals)
```

#### sets comprehension

```
#### 반복문
animals = {'cat', 'dog', 'fish'}
for idx, animal in enumerate(animals):
    print('#%d: %s' % (idx + 1, animal))
```

```
####확장 문법 - 반복문
from math import sqrt
nums = {int(sqrt(x)) for x in range(30)}
print(nums)  # "{0, 1, 2, 3, 4, 5}"
```

# 4.자료형(Container) - Tuples
- 수정이 불가능한 List. 수정 불가 를 제외하면 List와 똑 같다.
- Dictionary의 key로 이용될 수 있다.
- 값이 바뀌면 안되는 경우에 활용한다. (예: 함수 리턴값, 매개변수값)

```
# 튜플 생성
t = (5, 6)
print(type(t))    # Prints "<class 'tuple'>"
print(t)    # "(5, 6)"
print(t[0])       # "5"
```

```
# 삭제 시도
del t[1]
print(t)
```

```
# 변경 시도
t[1] = 4
```

```
# dictionary의 키로 사용되는 tuple
d = {(1, 0): 10, (2, 0): 20, (0, 1): 1}
print(d[(1, 0)])  # "10"
print(d[(0, 1)])  # "1"
```

### 데이터 타입 변환

```
stu = [["Geoffery Hinton", 95, 65, 90],
       ["Yann Lecun", 80, 85, 90],
       ["Yoshua Bengio", 92, 88, 84],
       ["Andrew Ng", 75, 85, 100]]

# list to dict
stu = [(s[0], s[1:]) for s in stu]
print(stu)
stu = dict(stu)
print(stu)

# dict to list
print(list(stu))
print(list(stu.keys()))
print(list(stu.values()))
print(list(stu.items()))
print([v for k, v in stu.items()])
```
