# 7.Lambda
- 사용법 - lambda [인자] : [표현식]
- 사용법 - lambda [인자] : [리턴값]

def와 동일하다. (머리 아프면 몰라도 된다)

보통 함수를 한줄로 간결하게 만든다.

return 없이 결과를 돌려준다.



```
# 일반 함수
def add0 (a, b):
    return (a + b)

print(add0(3, 4))
```

```
# lamda
add = lambda a, b: a+b
result = add(3, 4)
print(result)
```

```
pwr = lambda x: x ** 2
print(pwr(3))
```

```
def grade(my_score):
    if my_score >= 90:
        my_grade = 'A'
    elif my_score >= 80:
        my_grade = 'B'
    elif my_score >= 70:
        my_grade = 'C'
    elif my_score >= 60:
        my_grade = 'E'
    else:
        my_grade = 'F'

    return my_grade

func = lambda x: grade(x)
print(func(100))
print(grade(95))
```

```
stu01 = {
    "no": 1,
    "name": "Geoffery Hinton",
    "points": [92, 96, 98]
}
stu02 = {
    "no": 2,
    "name": "Yann Lecun",
    "points": [58, 64, 100]
}
stu03 = {
    "no": 3,
    "name": "Yoshua Bengio",
    "points": [88, 84, 92]
}
stu04 = {
    "no": 4,
    "name": "Andrew Ng",
    "points": [75, 85, 100]
}
stu = [stu01, stu02, stu03, stu04]
```

```
# 학생별 최고 점수를 보여줘 1. 한 학생만
def maxp(s):
    p = s['points']
    # print(p)
    return max(p)

maxp(stu01)
```

```
# 학생별 최고 점수를 보여줘: 전체 학생 대상으로
def maxp(stu):
    maxp = []
    for s in stu:
        p = s['points']
        # print(p)
        maxp.append(max(p))
    return maxp

maxp(stu)
```

```
# map, filter를 모른다면 9번을 하고 나서 공부하자

# 학생별 최고 점수를 보여줘: lamda를 쓰면
list(map(lambda x: max(x['points']), stu))
```

```
# map, filter를 모른다면 9번을 하고 나서 공부하자

# 이름 길이가 열 자리 초과인 것만
list(filter(lambda x: len(x['name']) > 10, stu))
```

```
# map, filter를 모른다면 9번을 하고 나서 공부하자

# 성적 60이하인 사람을 보여줘
# list(filter(lamda x: max(x['points']) <= 60, stu)
list(filter(lambda x: max(x['points']) <= 60, stu))
```

```
# 정렬하는데, 키는 points[0]으로 == (58, 75, 88, 92)
sorted(stu, key=lambda x: x['points'][0])
```

# 8.Decorator

\@ 을 붙여 함수 재사용을 간단히 할 수 있다.

```
def main_function():
    print ("MAIN FUNCTION START")

# 해당 문장을 출력하기 전과 후에 날짜와 시간을 출력하고 싶다면?
import datetime
def main_function():
    print (datetime.datetime.now())
    print ("MAIN FUNCTION START")
    print (datetime.datetime.now())

# 이런 함수가 100개라면?
def main_function_1():
    print ("MAIN FUNCTION 1 START")
def main_function_2():
    print ("MAIN FUNCTION 2 START")
# ..... X 100번..

# decorator를 사용하면 쉽다.
import datetime
def datetime_decorator(func):
    def decorated():
        print (datetime.datetime.now())
        func()
        print (datetime.datetime.now())
    return decorated

@datetime_decorator
def main_function_1():
    print ("MAIN FUNCTION 1 START")

@datetime_decorator
def main_function_2():
    print ("MAIN FUNCTION 2 START")

# ..... X 100번..

```

```
def decorator(origin):
    def wrapper():
        print('{} 함수 호출 전'.format(origin.__name__))
        return origin()
    return wrapper

def func_1():
    print('func_1 실행')


def func_2():
    print('func_2 실행')

f1 = decorator(func_1)
f2 = decorator(func_2)

f1()
f2()
```

```
def decorator(origin):
    def wrapper():
        print('{} 함수 호출 전'.format(origin.__name__))
        return origin()
    return wrapper

@decorator
def func_1():
    print('func_1 실행')

@decorator
def func_2():
    print('func_2 실행')

# f1 = decorator(func_1)
# f2 = decorator(func_2)

func_1()
func_2()
```

#9.Filter / Map/ Reduce

* filter(함수, 리스트)

특정 조건을 만족하는 요소만 남기고 필터링

* map

각 원소를 주어진 수식에 따라 변형하고 새 리스트를 만든다

* reduce:

차례대로 앞 2개의 원소를 가지고 연산. 연산의 결과가 다음 연산의 입력. 최종 출력은 하나의 값만 남는다.

```
def 짝수(n):
    return n%2 == 0

짝수(3)
```

```
nums = [1,2,3,6,8,15]
list (filter(짝수, nums))
```

```
# 람다로 더 줄이면
list (filter(lambda n:n%2==0, nums))
```

```
list( map(짝수, nums))
```

```
import functools
# 주어진 리스트를 전부 더하자.
functools.reduce( lambda x,y: x+y, nums)
```

```
# 주어진 리스트를 전부 곱하자.
functools.reduce( lambda x,y: x*y, nums)
```

### 실습: 함수, 람다, reduce

1. 주어진 숫자 리스트의 평균을 구하세요
2. 해당 숫자가 소수인지 아닌지 판별하세요
<br>소수: 1과 자기 자신으로만 나눠떨어지는 수.
<br>그러니까, 2부터 계속 자기 자신까지 값을 증가시키면서 나눠본
3. 2부터 해당 숫자 사이에 소수가 몇개인지 출력하세요


```
#1. 평균

nums = [1,2,3,6,8,15]

def mean1(l):
    sum = functools.reduce( lambda x,y:x+y, l)
    n = len(l)
    return sum/n

def mean2(l):
    sum = 0
    for i in l:
        sum += i
    return sum / len(l)

print (mean1(nums))
print (mean2(nums))
```

```
#2. 소수
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
        return True

print(is_prime(3))
print(is_prime(30))
print(is_prime(7))
```

```
#3. 주어진 숫자까지의 소수 갯수
def prime(n):
    count = 0
    for i in range(2, n+1): # 자기자신도 포함해서 세어야 하므로 n 이 아니라 n+1
        if n%i == 0:
            count += 1
    return count

print(prime(3))
print(prime(30))
print(prime(10))
```
