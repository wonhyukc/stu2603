### [S1] 1. 단일 if문 - 롤러코스터 키 제한 (기본 시연)

```python
height = 135
if height >= 130:
    print("범퍼카 탑승 가능! 즐거운 시간 되세요!")
```

### [S1] 2. 단일 if문 블록 묶음 - 은행 잔액 인출 시스템 (다중 출력 시연)

```python
balance = 50000
withdraw_amount = 30000

if balance >= withdraw_amount:
    print("현금을 인출합니다. 잠시만 기다려주세요.")
    balance -= withdraw_amount
    print("남은 잔액은", balance, "원입니다.")
```

### [S1] 3. if-else 양자택일 구조 - 합격 불합격 검증

```python
score = 65
if score >= 60:
    print("축하합니다! 합격입니다.")
else:
    print("아쉽습니다. 재시험 대상자입니다.")
```

### [S1] 4. if-elif-else 다중 분기 - 고속도로 요금소 할증 로직

```python
car_type = "승용차" # 값을 "화물차"나 "경차"로 바꿔서 시연해보세요

if car_type == "경차":
    print("통행료 50% 할인 대상입니다.")
elif car_type == "승용차":
    print("표준 통행료가 적용됩니다.")
elif car_type == "화물차":
    print("통행료 20% 할증이 적용됩니다.")
else:
    print("알 수 없는 차종입니다. 안내소로 문의하세요.")
```

### [S2] 5. and 연산자 엄격성 시뮬레이션 - 청소년 관람불가 인증 시스템

```python
age = 20
has_id_card = False # 신분증을 안 가져왔다면 어떻게 될까?

if age >= 19 and has_id_card == True:
    print("입장을 허가합니다. 즐거운 관람 되십시오.")
else:
    print("나이 조건 미달이거나 신분증을 지참하지 않으셨습니다. 입장 불가!")
```

### [S2] 6. or 연산자의 너그러움 판독 - 영수증 OR VIP 주차 무료

```python
is_vip_member = False
shopping_amount = 65000

if is_vip_member or shopping_amount >= 50000:
    print("고객님, 1시간 주차 요금 무료 처리되었습니다.")
else:
    print("요금 3,000원을 결제해 주세요.")
```

### [S2] 7. not 연산자의 마법 반전 - 방패 미장착 데미지 결산

### [S2] 8. 중첩 if문의 깊이 탐색 - 온라인 2단계 비밀번호 인증

```python
username = "admin"
password_input = "1234"

if username == "admin":
    print("관리자 계정이 확인되었습니다. 1차 관문 통과.")
    if password_input == "1234":
        print("비밀번호 완전 일치! 시스템에 접속합니다.")
    else:
        print("비밀번호가 틀렸습니다. 경고!")
else:
    print("등록되지 않은 사용자입니다.")
```

### [S3] 10. 단일 학점 계산기 (어리석은 8500점 입력 시나리오 - Edge Case)

```python
# 강사가 일부러 8500 을 입력해서 에러 없이 A가 나오는 모순 상황을 시연하는 셀
score = int(input("말도 안되는 8500점을 입력해보세요: "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"최종결과: {score}점, 등급은 {grade}입니다.")
```

### [S3] 11. 최종 AI 협업본: 강력한 Validation 방어 로직이 적용된 학점 계산기

```python
# 최종 버전 (경계값 검사 포함 방어형 코드)
score_str = input("최종 점수를 입력하세요 (예외 처리 완비): ")
score = int(score_str)

if score < 0 or score > 100:
    print("⚠ 치명적 오류: 점수는 사람이 맞을 수 있는 0점에서 100점 사이여야 합니다!!!")
else:
    print("점수가 정상 범위 내에 있습니다. 심사를 시작합니다...")
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    else:
        grade = "F"
        
    print(f"최종 판독 결과: {score}점 획득으로 '{grade}' 학점을 받으셨습니다.")
```
