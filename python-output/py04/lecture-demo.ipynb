## 4주차 강사용 시연 데모 모음

### [S1] 특정 점수 이상 통과 (단일 if)

```python
score = 85
if score >= 80:
    print("합격입니다! 축하합니다!")
```

### [S1] if-else 분기

```python
score = 70
if score >= 80:
    print("합격입니다!")
else:
    print("아쉽네요, 다음 기회에!")
```

### [S1] if-elif-else 다중 분기

```python
score = 85
if score >= 90:
    print("최우수 A등급")
elif score >= 80:
    print("우수 B등급")
else:
    print("조금 더 분발 C등급")
```

### [S2] and 연산자 실습

```python
height = 135
has_disease = False

if height >= 130 and has_disease == False:
    print("롤러코스터 탑승을 환영합니다!")
else:
    print("죄송하지만 탑승하실 수 없습니다.")
```

### [S2] or 연산자 실습

```python
is_vip = False
purchase_amount = 120000

if is_vip == True or purchase_amount >= 100000:
    print("10% 파격 할인을 해드립니다!")
```

### [S2] not 연산자 실습

```python
is_member = False
if not is_member:
    print("신규 회원 가입을 진행하시겠습니까?")
```

### [S2] 중첩 if문 (Nested if)

```python
age = 25
if age >= 18:
    print("성인 구역에 오신 것을 환영합니다.")
    # 이 안에서 또 if문 시작
    if age >= 20: 
        print("맥주를 주문하실 수 있습니다!")
```

### [S3] 라이브 코딩 및 AI 디버깅

```python
score_input = input("학생의 점수를 입력하세요 (0~100): ")
score = int(score_input)

if score < 0 or score > 100:
    print("경고: 점수는 0에서 100 사이여야 합니다!!")
elif score >= 90:
    grade = "A"
    print(f"학점은 {grade}입니다!")
elif score >= 80:
    grade = "B"
    print(f"학점은 {grade}입니다!")
elif score >= 70:
    grade = "C"
    print(f"학점은 {grade}입니다!")
else:
    grade = "F"
    print(f"학점은 {grade}입니다!")
```
