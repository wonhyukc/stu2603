### 산술 연산자


```python
# 사칙연산
print(10 + 3)
print(10 - 3)
print(10 * 3)
print(10 / 3)
```

```python
# 특수 연산자 (몫, 나머지, 거듭제곱)
print(10 // 3)
print(10 % 3)
print(2 ** 3)
print(3 ** 2)
print(2 ** 10)
```

**[추가 데모] 짝수/홀수 판별에 나머지 연산자 활용**


```python
number = 7
print("7을 2로 나눈 나머지:", number % 2) # 나머지가 1이면 홀수, 0이면 짝수

number2 = 8
print("8을 2로 나눈 나머지:", number2 % 2)
```

## Section 3: input() + 타입 변환 + 에러
### input() 기본 사용법


```python
name = input("이름을 입력하세요: ")
print("안녕하세요,", name)
```

```python
user_input = input("숫자를 입력하세요: ")
print("입력된 값의 타입:", type(user_input))
```

```python
# 에러가 발생하는 코드 (입력값은 무조건 문자열이라 10을 더할 수 없음)
# num = input("숫자를 입력하세요: ")
# result = num + 10
```

### 타입 변환 (int, float, str)


```python
num_str = "42"
num_int = int(num_str)
print(num_int + 10)
print("변환 후 타입:", type(num_int))
```

```python
age = int(input("나이를 입력하세요: "))
print("내년에는", age + 1, "살이 됩니다.")
```

```python
height_str = "175.5"
height_float = float(height_str)
print(height_float + 2.0)
print("변환 후 타입:", type(height_float))
```

```python
score = float(input("점수를 입력하세요: "))
print("입력한 점수:", score)
```

```python
age = 20
message = "현재 나이는 " + str(age) + "살 입니다."
print(message)
```
