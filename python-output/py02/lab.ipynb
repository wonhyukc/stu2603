## Section 1: print() 완전 정복

### 1-1. print() 기본 출력

아래 코드를 실행해 보세요. 그다음 직접 원하는 문장으로 바꿔서 출력해 보세요.

```python
# 문자열 출력
print("안녕하세요, 파이썬!")

# 숫자 출력
print(42)
print(3.14)
```

### 1-2. 콤마(,) vs 더하기(+) 비교

두 방식의 차이를 눈으로 확인하세요.
- **콤마**: 값 사이에 공백 자동 삽입, 타입이 달라도 OK
- **더하기**: 공백 없이 바짝 붙음, 문자열끼리만 가능

```python
# 콤마로 여러 값 출력 — 자동 공백
print("나는", "파이썬을", "배우고 있다")

# 콤마로 문자열 + 숫자 출력 — 에러 없음
나이 = 20
print("나의 나이는", 나이, "살입니다.")
```

```python
# 더하기로 문자열 연결 — 공백 없이 붙음
print("안녕" + "하세요")

# 더하기로 문자 반복
print("=" * 30)
print("하하" * 3)
```

```python
# 더하기로 문자열 + 숫자 시도 — 에러 발생!
# 아래 코드를 실행해서 TypeError 메시지를 확인하세요.
나이 = 20
print("나의 나이는 " + 나이 + "살입니다.")
```

### 1-3. end= 옵션과 sep= 옵션

- `end=` : 출력 끝에 붙는 문자를 바꿉니다 (기본값: 줄바꿈 `"\n"`)
- `sep=` : 콤마로 나열한 값들 사이의 구분자를 바꿉니다 (기본값: 공백 `" "`)

```python
# end= 옵션
print("로딩 중", end="...")
print("완료!")

# 줄바꿈 없이 이어서 출력
print("A", end="")
print("B", end="")
print("C")
```

```python
# sep= 옵션
print("2026", "03", "07", sep="-")         # 날짜 형식
print("사과", "바나나", "딸기", sep=", ")   # 목록 형식
print("A", "B", "C", sep="→")             # 화살표 구분

# end=와 sep= 동시 사용
print("사과", "바나나", sep=" | ", end="!\n")
```

## Section 2: 4가지 기본 타입 + 산술 연산자

### 2-1. 4가지 기본 타입 확인

파이썬의 기본 타입 4가지:
- `str` — 문자열 (따옴표로 감싼 것)
- `int` — 정수 (소수점 없는 숫자)
- `float` — 부동소수점 (소수점 있는 숫자)
- `bool` — 불리언 (`True` / `False`)

```python
# str (문자열)
이름 = "정원혁"
print(이름)

# 숫자처럼 보여도 따옴표 안에 있으면 str!
숫자처럼_보이는_문자열 = "42"
진짜_숫자 = 42
print("42" + "1")  # 421  (문자열 연결)
print(42 + 1)      # 43   (숫자 덧셈)
```

```python
# int (정수)
나이 = 20
온도 = -5
print(나이, 온도)

# float (부동소수점)
키 = 175.5
원주율 = 3.14159
print(키, 원주율)

# 주의: 정수 나눗셈의 결과는 float!
print(10 / 2)   # 5.0  (int가 아님)
```

```python
# bool (불리언)
로그인_상태 = True
시험_합격 = False
print(로그인_상태, 시험_합격)

# 비교 연산의 결과도 bool
print(10 > 5)    # True
print(10 < 5)    # False
print(10 == 10)  # True
```

### 2-2. type() 함수로 타입 확인하기

`type()` 함수를 사용하면 어떤 데이터의 타입인지 바로 확인할 수 있습니다.  
헷갈릴 때마다 써보세요.

```python
# 각 타입 확인
print(type("안녕하세요"))   # <class 'str'>
print(type(42))             # <class 'int'>
print(type(3.14))           # <class 'float'>
print(type(True))           # <class 'bool'>
```

```python
# 직접 확인해 보기
결과 = 10 / 2
print(결과)          # 5.0
print(type(결과))    # <class 'float'>

# 아래 변수들의 타입을 type()으로 직접 확인해 보세요.
x = "100"
y = 100
z = True
# print(type(x))  # 주석 해제해서 확인해 보세요
# print(type(y))
# print(type(z))
```

### 2-3. 산술 연산자 7종

| 연산자 | 이름 | 예시 (`a=17, b=5`) | 결과 |
|:------:|:-----|:-------------------|:----:|
| `+` | 더하기 | `a + b` | `22` |
| `-` | 빼기 | `a - b` | `12` |
| `*` | 곱하기 | `a * b` | `85` |
| `/` | 나누기 | `a / b` | `3.4` |
| `//` | 정수 나누기 (몫) | `a // b` | `3` |
| `%` | 나머지 | `a % b` | `2` |
| `**` | 거듭제곱 | `a ** b` | `1419857` |

```python
a = 17
b = 5

print("더하기:", a + b)
print("빼기:", a - b)
print("곱하기:", a * b)
print("나누기:", a / b)          # 결과가 float!
print("정수 나누기(몫):", a // b)
print("나머지:", a % b)
print("거듭제곱:", a ** b)
```

```python
# % 활용: 짝수/홀수 판별
숫자 = 15
print(숫자, "% 2 =", 숫자 % 2)   # 1이면 홀수, 0이면 짝수

숫자2 = 14
print(숫자2, "% 2 =", 숫자2 % 2)

# ** 활용: 2의 거듭제곱
print("2의 10제곱:", 2 ** 10)   # 1024

# 연산자 우선순위 — 괄호로 명확하게
print(2 + 3 * 4)     # 14 (곱하기 먼저)
print((2 + 3) * 4)   # 20 (괄호 먼저)
```

## Section 3: input() + 타입 변환 + 에러

### 3-1. input() 함수: 사용자 입력 받기

`input()` 함수는 실행 중에 잠깐 멈추고 사용자 입력을 기다립니다.  
사용자가 Enter를 누르면 입력한 값을 반환합니다.  

> **중요**: `input()`은 입력값을 **항상 str(문자열)로 반환**합니다.

```python
# 이름 입력받아 출력
이름 = input("이름을 입력하세요: ")
print("안녕하세요,", 이름, "님!")
```

```python
# input()의 반환값 타입 확인
# 숫자를 입력해도 결과는 str!
나이 = input("나이를 입력하세요: ")
print("입력한 값:", 나이)
print("타입:", type(나이))   # <class 'str'>
```

### 3-2. 타입 변환: int(), float(), str()

- `int()` — 정수로 변환
- `float()` — 부동소수점으로 변환
- `str()` — 문자열로 변환

```python
# int()로 변환 후 계산
나이 = int(input("나이를 입력하세요: "))
다음_나이 = 나이 + 1
print("내년에는", 다음_나이, "살입니다.")
print("변환 후 타입:", type(나이))   # <class 'int'>
```

```python
# float()로 변환
키 = float(input("키를 입력하세요 (cm): "))
print("키는", 키, "cm입니다.")
print("변환 후 타입:", type(키))   # <class 'float'>
```

```python
# str()로 변환 — 숫자를 문자열에 더하기(+)로 연결할 때 필요
나이 = 20
문장 = "나의 나이는 " + str(나이) + "살입니다."
print(문장)

# 변환 불가능한 경우 — ValueError 확인
# 아래 주석을 해제하고 실행해서 에러 메시지를 확인해 보세요.
# print(int("안녕"))
# print(int("3.14"))   # 소수점 있는 문자열은 int()로 직접 변환 불가
print(int(float("3.14")))   # float 거쳐서 int로 변환
```

### 3-3. SyntaxError vs TypeError 비교

| 에러 | 발생 시점 | 원인 |
|:-----|:---------|:-----|
| `SyntaxError` | 실행 전 | 파이썬 문법을 어긴 경우 |
| `TypeError` | 실행 중 | 타입이 맞지 않는 연산을 시도한 경우 |

아래 셀들을 실행해서 두 에러 메시지의 차이를 직접 확인해 보세요.

```python
# SyntaxError 유발 — 따옴표를 닫지 않음
# 에러 메시지를 읽고 무엇이 문제인지 파악하세요.
# 그다음 AI에게 에러 메시지를 붙여넣어 질문해 보세요.
print("안녕하세요)
```

```python
# TypeError 유발 — str + int 연산 시도
# 실행 후 에러 메시지를 읽고 어떤 타입 충돌인지 확인하세요.
나이 = input("나이를 입력하세요: ")   # 반환값은 str
다음_나이 = 나이 + 1                  # str + int → TypeError!
```
