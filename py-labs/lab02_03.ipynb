```
# return을 사용하면 함수 밖으로 값을 전달할 수 있다.
a = 1
def vartest(a):
    a = a +1
    return a

a = vartest(a)
print(a)
```

```
# 다음도 가능하지만 바람직하지 않다.
# global은 가급적 사용하지 말자. 함수의 독립성을 해친다.
a = 1
def vartest():
    global a
    a = a+1

vartest()
print(a)
```

#### seek & tell

```
f = open('./new_file.txt', 'r')
print('파일 현재 위치:', f.tell())
print(f.read())
print('파일 현재 위치:', f.tell())
print()

f.seek(0) # 처음으로

print('파일 현재 위치:', f.tell())
print(f.readline(), end='')
print('파일 현재 위치:', f.tell())
f.close()
```

```
def search_path(path):
    print("path - {0} ####################".format(path))
    files = os.listdir(path)
    for file_name in files:
        full = os.path.join(path, file_name)
        print(full)

search_path('./')
```

```
def search_path(path):
    files = os.listdir(path)
    for file_name in files:
        full = os.path.join(path, file_name)
        print(full)
        if (os.path.isdir(full)):
            search_path(full)

search_path('./')
```

```
import os

for (path, dir, files) in os.walk("./"):
    for filename in files:
        ext = os.path.splitext(filename)[-1]
        print("%s/%s" % (path, filename))
```
