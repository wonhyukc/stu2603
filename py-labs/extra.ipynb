## 3.가변인자

함수에 매개변수가 몇개나 올지 모를때

\*args: 임의 갯수의 위치 arguments

\**kwarg: 임의 갯수의 키워드 arguments

```
# 일반적 함수의 매개변수
# 2~4명의 주자로 이루어진 달리기 대회 랭킹을 보여주는 함수
def save_ranking(first, second, third=None, fourth=None):
    rank = {}
    rank[1], rank[2] = first, second
    rank[3] = third if third is not None else 'Nobody'
    rank[4] = fourth if fourth is not None else 'Nobody'
    print(rank)

# positional arguments 2개 전달
save_ranking('ming', 'alice')
# positional arguments 2개와 keyword argument 1개 전달
save_ranking('alice', 'ming', third='mike')
# positional arguments 2개와 keyword arguments 2개 전달 (단, 하나는 positional argument 형태로 전달)
save_ranking('alice', 'ming', 'mike', fourth='jim')
```

```
# 몇명이 달릴지 모른다면?
# *args로 위치 매개변수를 전달.
def save_ranking(*args):
    print(args)
save_ranking('ming', 'alice', 'tom', 'wilson', 'roy')
# ('ming', 'alice', 'tom', 'wilson', 'roy')
```

```
# **kwargs로 키워드매개변수를 전달
def save_ranking(**kwargs):
    print(kwargs)
save_ranking(first='ming', second='alice', fourth='wilson', third='tom', fifth='roy')
# {'first': 'ming', 'second': 'alice', 'fourth': 'wilson', 'third': 'tom', 'fifth': 'roy'}
```

\@ 을 붙여 함수 재사용을 간단히 할 수 있다.

```
def main_function():
    print ("MAIN FUNCTION START")

# 해당 문장을 출력하기 전과 후에 날짜와 시간을 출력하고 싶다면?
import datetime
def main_function():
    print (datetime.datetime.now())
    print ("MAIN FUNCTION START")
    print (datetime.datetime.now())

# 이런 함수가 100개라면?
def main_function_1():
    print ("MAIN FUNCTION 1 START")
def main_function_2():
    print ("MAIN FUNCTION 2 START")
# ..... X 100번..

# decorator를 사용하면 쉽다.
import datetime
def datetime_decorator(func):
    def decorated():
        print (datetime.datetime.now())
        func()
        print (datetime.datetime.now())
    return decorated

@datetime_decorator
def main_function_1():
    print ("MAIN FUNCTION 1 START")

@datetime_decorator
def main_function_2():
    print ("MAIN FUNCTION 2 START")

# ..... X 100번..

```

```
def decorator(origin):
    def wrapper():
        print('{} 함수 호출 전'.format(origin.__name__))
        return origin()
    return wrapper

def func_1():
    print('func_1 실행')


def func_2():
    print('func_2 실행')

f1 = decorator(func_1)
f2 = decorator(func_2)

f1()
f2()
```

### 실습: 함수, 람다, reduce

1. 주어진 숫자 리스트의 평균을 구하세요
2. 해당 숫자가 소수인지 아닌지 판별하세요
<br>소수: 1과 자기 자신으로만 나눠떨어지는 수.
<br>그러니까, 2부터 계속 자기 자신까지 값을 증가시키면서 나눠본
3. 2부터 해당 숫자 사이에 소수가 몇개인지 출력하세요


```
#1. 평균

nums = [1,2,3,6,8,15]

def mean1(l):
    sum = functools.reduce( lambda x,y:x+y, l)
    n = len(l)
    return sum/n

def mean2(l):
    sum = 0
    for i in l:
        sum += i
    return sum / len(l)

print (mean1(nums))
print (mean2(nums))
```

```
#2. 소수
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
        return True

print(is_prime(3))
print(is_prime(30))
print(is_prime(7))
```

```
#3. 주어진 숫자까지의 소수 갯수
def prime(n):
    count = 0
    for i in range(2, n+1): # 자기자신도 포함해서 세어야 하므로 n 이 아니라 n+1
        if n%i == 0:
            count += 1
    return count

print(prime(3))
print(prime(30))
print(prime(10))
```

# 계산기

```
# 들어오는 값을 더하는 더하기 함수를 만들자. 계산 된 값은 이후 계산에 그대로 반영되어야 한다.
r = 0

def add(num):
    global r
    r += num
    return r

print(add(3))
print(add(4))
```

```
# 계산기가 두 개 필요하다. add1, add2함수를 만들자.
r1, r2 = 0, 0

def add1(num):
    global r1
    r1 += num
    return r1

def add2(num):
    global r2
    r2 += num
    return r2

print(add1(3))
print(add1(4))

print(add2(7))
print(add2(4))
```

```
# 곱하기 / 나누기 만들기
class Calculator:
    def setdata(self, x, y): # 첫번째 인자는 항상 자기 자신이다. 그래서 self라는 이름을 관용적으로 사용한다.
        self.x = x
        self.y = y
    def add(self):
        r = self.x + self.y
        return r
    def sub(self):
        r = self.x - self.y
        return r
    def mul(self):
        r = self.x * self.y
        return r
    def div(self):
        r = self.x / self.y
        return

a = Calculator()
a.setdata(4, 2)
print(a.add())
print(a.mul())
print(a.div())
```

## Constructor

```
# setdata() 하지 않고 바로 add() 를 호출하고 싶다.

a = Calculator()
print(a.add())
```

```
# constructor는 객체 생성시 자동으로 호출되는 method이다.
class Calculator:
    def __init__ (self, x, y):
        self.x = x
        self.y = y
    def setdata(self, x, y): # 첫번째 인자는 항상 자기 자신이다. 그래서 self라는 이름을 관용적으로 사용한다.
        self.x = x
        self.y = y
    def add(self):
        r = self.x + self.y
        return r
    def sub(self):
        r = self.x - self.y
        return r
    def mul(self):
        r = self.x * self.y
        return r
    def div(self):
        r = self.x / self.y
        return

a = Calculator(4, 2)
print(a.add())
print(a.sub())
```

# 상속

```
# 원래 클래스에 부가적 기능을 더하고 싶을 때 / 변경을 가하고 싶을 때
# 원래 클래스는 건드리지 않고

class MoreCalculator(Calculator):
    pass

a = MoreCalculator(4,2)
a.add()
```

```
# 제곱 기능을 추가해보자.
class MoreCalculator(Calculator):
    def pow(self):
        r = self.x ** self.y
        return r

a = MoreCalculator(4,3)
a.pow()
```

# Method Overriding

```
# 0으로 나누면 당연히 오류
a = Calculator(4,0)
a.div()
```

```
# 클래스 상속
# Method Override
class safeCalculator(Calculator):
    def div(self):
        if self.y == 0:
            return 0
        else:
            return self.x / self.y

a = safeCalculator(4, 0)
a.div()
```
