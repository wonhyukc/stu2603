## 3.가변인자

함수에 매개변수가 몇개나 올지 모를때

\*args: 임의 갯수의 위치 arguments

\**kwarg: 임의 갯수의 키워드 arguments

```
# 일반적 함수의 매개변수
# 2~4명의 주자로 이루어진 달리기 대회 랭킹을 보여주는 함수
def save_ranking(first, second, third=None, fourth=None):
    rank = {}
    rank[1], rank[2] = first, second
    rank[3] = third if third is not None else 'Nobody'
    rank[4] = fourth if fourth is not None else 'Nobody'
    print(rank)

# positional arguments 2개 전달
save_ranking('ming', 'alice')
# positional arguments 2개와 keyword argument 1개 전달
save_ranking('alice', 'ming', third='mike')
# positional arguments 2개와 keyword arguments 2개 전달 (단, 하나는 positional argument 형태로 전달)
save_ranking('alice', 'ming', 'mike', fourth='jim')
```

```
# 몇명이 달릴지 모른다면?
# *args로 위치 매개변수를 전달.
def save_ranking(*args):
    print(args)
save_ranking('ming', 'alice', 'tom', 'wilson', 'roy')
# ('ming', 'alice', 'tom', 'wilson', 'roy')
```

```
# **kwargs로 키워드매개변수를 전달
def save_ranking(**kwargs):
    print(kwargs)
save_ranking(first='ming', second='alice', fourth='wilson', third='tom', fifth='roy')
# {'first': 'ming', 'second': 'alice', 'fourth': 'wilson', 'third': 'tom', 'fifth': 'roy'}
```

## 4.Unpacking: container 자료형을

```
numbers = 1, 2, 3, 4, 5  # 패킹
a, b, c, d, e = numbers  # 언패킹
# 언패킹할 때, 즉 시퀀스의 요소를 변수 시퀀스에 나눠 대입할 때는 두 시퀀스의 길이가 일치해야 한다.
print(numbers)
print(a)
print(b)
```

```
a, b, c, d, _ = numbers  # ❷ 필요 없는 요소를 _ 변수에 대입
# ❶처럼 길이가 일치하지 않으면 오류가 발생한다. 대입할 필요가 없는 요소는 ❷와 같이 자리를 메우는 변수에 대입하면 된다. 자리를 메우는 변수의 이름은 관례적으로 밑줄 기호 하나(_)로 한다.

# 남은 요소 대입받기
# 대입문에서 좌변의 변수 중 하나에 별 기호(*)를 붙이면, 남은 요소 전체를 리스트에 담아 대입한다.

a = numbers[0]            # a = 1
b = numbers[1]            # b = 2
rest = numbers[2:]        # rest = (3, 4, 5)
print(a, b, rest)
```

```
a, b, *rest = numbers     # 1, 2를 제외한 나머지를 rest에 대입
print(a, b, rest)
```

```
*rest, c, d, e = numbers  # 3, 4, 5를 제외한 나머지를 rest에 대입
print(rest)
```

```
a, *rest, e = numbers     # 1, 5를 제외한 나머지를 rest에 대입
print(rest)
```

# 계산기

```
# 들어오는 값을 더하는 더하기 함수를 만들자. 계산 된 값은 이후 계산에 그대로 반영되어야 한다.
r = 0

def add(num):
    global r
    r += num
    return r

print(add(3))
print(add(4))
```

```
# 계산기가 두 개 필요하다. add1, add2함수를 만들자.
r1, r2 = 0, 0

def add1(num):
    global r1
    r1 += num
    return r1

def add2(num):
    global r2
    r2 += num
    return r2

print(add1(3))
print(add1(4))

print(add2(7))
print(add2(4))
```

## Constructor

```
# setdata() 하지 않고 바로 add() 를 호출하고 싶다.

a = Calculator()
print(a.add())
```

```
# constructor는 객체 생성시 자동으로 호출되는 method이다.
class Calculator:
    def __init__ (self, x, y):
        self.x = x
        self.y = y
    def setdata(self, x, y): # 첫번째 인자는 항상 자기 자신이다. 그래서 self라는 이름을 관용적으로 사용한다.
        self.x = x
        self.y = y
    def add(self):
        r = self.x + self.y
        return r
    def sub(self):
        r = self.x - self.y
        return r
    def mul(self):
        r = self.x * self.y
        return r
    def div(self):
        r = self.x / self.y
        return

a = Calculator(4, 2)
print(a.add())
print(a.sub())
```

# Method Overriding

```
# 0으로 나누면 당연히 오류
a = Calculator(4,0)
a.div()
```

```
# 클래스 상속
# Method Override
class safeCalculator(Calculator):
    def div(self):
        if self.y == 0:
            return 0
        else:
            return self.x / self.y

a = safeCalculator(4, 0)
a.div()
```
